```python
import matplotlib.pyplot as plt
import math as m
import time

xx=[]
yy=[]

for i in range(0, 181):
    t = m.radians(i*2)
    x = 16*m.sin(t)**3
    y = 13*m.cos(t) - 5*m.cos(2*t) - 2*m.cos(3*t) - m.cos(4*t)
    xx.append(x)
    yy.append(y)


plt.plot(xx, yy, 'r-')
plt.axis([-20, 20, -20, 20])
plt.show()
```
